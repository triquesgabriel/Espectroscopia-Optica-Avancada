### Absorbância Clorofina


```python
# pd.read_csv('Abs_clorofila.txt', sep='\t', skiprows = 17)
# !head -n 17 Abs_clorofila.txt

Abs_cloro_df = pd.read_csv('Abs_clorofila.txt',
                 sep=r'\s+',
                 skiprows = 17,
                 decimal = ',',
                 names = ['Comprimento de onda', 'Abs'],
                 skipfooter=1,
                 engine='python')

Abs_cloro_df.plot(x = 'Comprimento de onda', y = 'Abs')


plt.title('Espectro de Absorbância - Clorofila')
plt.xlabel('Comprimento de Onda (nm)')
plt.ylabel('Absorbância')
plt.xlim(390, 750)
plt.ylim(0, 2)
plt.legend()

# Abs_cloro_df.plot.scatter(x = 'Comprimento de onda', y = 'Abs')
```
